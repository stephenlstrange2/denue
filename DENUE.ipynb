```python
import sys
from types import ModuleType, FunctionType
from gc import get_referents

# Custom objects know their class.
# Function objects seem to know way too much, including modules.
# Exclude modules as well.
BLACKLIST = type, ModuleType, FunctionType


def getsize(obj):
    """sum size of object & members."""
    if isinstance(obj, BLACKLIST):
        raise TypeError('getsize() does not take argument of type: '+ str(type(obj)))
    seen_ids = set()
    size = 0
    objects = [obj]
    while objects:
        need_referents = []
        for obj in objects:
            if not isinstance(obj, BLACKLIST) and id(obj) not in seen_ids:
                seen_ids.add(id(obj))
                size += sys.getsizeof(obj)
                need_referents.append(obj)
        objects = get_referents(*need_referents)
    return size
```

```python
def consultaBuscarDENUE(condicion,latitud,longitud,distancia,token="0f598ede-496c-4c6e-8f0c-10cfb207e2a5"):
    r = requests.get("https://www.inegi.org.mx/app/api/denue/v1/consulta/Buscar/"+condicion+"/"+latitud+","+longitud+"/"+distancia+"/"+token)
    return r.json()

def consultaFichaDENUE(Id,token="0f598ede-496c-4c6e-8f0c-10cfb207e2a5"):
    r = requests.get("https://www.inegi.org.mx/app/api/denue/v1/consulta/ficha/"+Id+"/"+token)
    return r.json()

def consultaNombreDENUE(nombre,entidad,registroInicial,registroFinal,token="0f598ede-496c-4c6e-8f0c-10cfb207e2a5"):
    r = requests.get("https://www.inegi.org.mx/app/api/denue/v1/consulta/Nombre/"+nombre+"/"+entidad+"/"+registroInicial+"/"+registroFinal+"/"+token)
    return r.json()

def consultaBuscarAreaActDENUE(Entidad,Municipio,Localidad,AGEB,Manzana,Sector,Subsector,Rama,Clase,Nombre,registroInicial,registroFinal,Id,token="0f598ede-496c-4c6e-8f0c-10cfb207e2a5"):
    r = requests.get("https://www.inegi.org.mx/app/api/denue/v1/consulta/BuscarAreaAct/"+Entidad+"/"+Municipio+"/"+Localidad+"/"+AGEB+"/"+Manzana+"/"+Sector+"/"+Subsector+"/"+Rama+"/"+Clase+"/"+Nombre+"/"+registroInicial+"/"+registroFinal+"/"+Id+"/"+token)
    return r.json()

def consultaCuantificarDENUE(Actividad,Area = "0", Estrato ="0",token="0f598ede-496c-4c6e-8f0c-10cfb207e2a5"):
    r = requests.get("https://www.inegi.org.mx/app/api/denue/v1/consulta/Cuantificar/"+Actividad+"/"+Area+"/"+Estrato+"/"+token)
    return r.json()
```
